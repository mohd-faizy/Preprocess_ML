#  Outlier

### What is an outlier?

In the context of machine learning, an outlier refers to a data point that `differs significantly` from other data points in the dataset. Outliers can be present in the input features or in the target variable.

Outliers can affect the `accuracy` of machine learning models, as they can `skew` the results and create misleading patterns. Therefore, it's important to identify and handle outliers appropriately.

In terms of a chart, an outlier would be a data point that falls significantly outside the range of the other data points in the chart. For example, if you have a `scatter plot` of two variables, an outlier would be a point that is far away from the other points in the plot. Similarly, in a box plot, an outlier would be a point that falls outside the `whiskers` of the box.

**Some more points to discuss about outliers in machine learning**:
  
- Outliers can have a **significant impact** on various machine learning tasks, including `regression`, `clustering`, and `classification`.
  
- Outliers can affect the statistical properties of a dataset, such as the
    - `mean`
    - `standard deviation`
    - and `correlation`, and can distort the shape of probability distributions.
  
- Identifying outliers can involve various techniques, such as `visual inspection`, `statistical tests`, and `machine learning algorithms` that are specifically designed for outlier detection.
  
- Handling outliers can involve various strategies, such as `removing` them from the dataset, `replacing` them with more typical values, or `modifying` the model to be more robust to outliers.
  
- Outliers can also **provide valuable insights into the problem** being solved, such as identifying `rare events` or revealing underlying patterns that are not visible in  the majority of the data.
  
- Outliers can be especially important in `anomaly detection` tasks, where the goal is to identify instances that are significantly different from the norm, and can be used to train and evaluate such models.
  
- Outliers can also be affected by the choice of machine learning algorithm, as some algorithms are more sensitive to outliers than others, and may require additional preprocessing steps to handle them effectively.

### **Most commonly used methods for identifying outliers**:

| Criteria | Method | Explanation |
| --- | --- | --- |
| Distance-based | Z-score | Points with a z-score greater than a certain threshold (e.g., 3), Data point that falls outside of 3 standard deviations. we can use a z score and if the z score falls outside of 2 standard deviation|
|  | Tukey method | Points that fall outside the range defined by Q1 - 1.5*IQR and Q3 + 1.5*IQR |
|  | Box plot | Points that fall outside the whiskers of the box plot |
|  | Interquartile range (IQR) method | Data point that falls outside of 1.5 times of an interquartile range above the 3rd quartile and below the 1st quartile |
|  | Hampel identifier | A robust method based on the median and median absolute deviation (MAD) of the data |
| Density-based | DBSCAN | Points that have low density compared to their neighbors are considered outliers |
|  | Local Outlier Factor (LOF) | Points that have a substantially lower density than their neighbors are considered outliers |
| Clustering-based | k-means | Points that are far away from any cluster centroid are considered outliers |
| Supervised | Regression-based models | Points that deviate significantly from the expected values predicted by the model are considered outliers |
|  | Decision trees | Points that fall outside the boundaries defined by the decision tree are considered outliers |
| Other | Cook's distance | Points that have a large effect on the regression line in linear regression analysis |
|  | Grubbs' test | Points that fall outside the confidence interval of the normal distribution |
|  | Modified Z-score | A modified version of the Z-score method that uses the median and median absolute deviation (MAD) to identify outliers |


### **What is the reason for an outlier to exists in a dataset?**

Some possible reasons for the existence of outliers are:

- **Data entry errors**: Mistakes in data collection or recording can lead to outliers in a dataset.
- **Measurement errors**: Errors in measurement instruments or methods can result in extreme values that do not reflect the true value of the variable.
- **Natural variation**: In some cases, outliers may simply be the result of natural variation in the data.
- **Skewed distribution**: If the distribution of a variable is highly skewed, it is more likely to have outliers.
- **Data processing errors**: Mistakes made during data cleaning or processing can introduce outliers in a dataset.
- **Sampling errors**: If a sample is not representative of the population, outliers may occur.
- **Extreme events**: Outliers can arise due to extreme or unusual events that do not follow the normal pattern of the data.

### **What are the impacts of having outliers in a dataset?**

- **Distorted measures of central tendency**: Outliers can significantly affect the mean, causing it to be pulled in the direction of the outlier. This can distort measures of central tendency, such as the `mean` and `median`, and make them less representative of the typical values in the dataset.
  
- **Increased variability**: Outliers can increase the variability or spread of the data, making it difficult to accurately estimate the true range of values in the dataset.
  
- **Biased statistical inference**: Outliers can lead to `biased statistical inference`, as they can affect the `estimates of parameters`, such as `regression coefficients`, and make them **less accurate** or **meaningful**.
  
- **Reduced predictive power**: Outliers can also **reduce the predictive power** of statistical models, as they can cause the model to fit the data poorly or overfit the data.
- **Difficulty in detecting patterns**: Outliers can make it **difficult to detect patterns or relationships** in the data, as they `can dominate the analysis` and obscure any underlying trends or associations.
  
- **Reduced data quality**: Outliers can reduce the overall quality of the data, as they may indicate errors or anomalies in the data collection process or the underlying system being studied.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data with some outliers
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Adding the outliers
x_outliers = np.random.normal(5, 1, 3)
y_outliers = np.random.normal(5, 1, 3)

# adding the outliers to our data
x = np.concatenate((x, x_outliers))
y = np.concatenate((y, y_outliers))

# Create scatter plot
plt.scatter(x, y)

# Calculate mean and standard deviation
x_mean = np.mean(x)
y_mean = np.mean(y)
x_std = np.std(x)
y_std = np.std(y)

# Set threshold for identifying outliers
threshold = 3

# Identify outliers
x_outliers = [i for i, j in enumerate(x) if abs(j - x_mean) > threshold * x_std]
y_outliers = [i for i, j in enumerate(y) if abs(j - y_mean) > threshold * y_std]
outliers = list(set(x_outliers + y_outliers))

# Highlight outliers in red
plt.scatter(x[outliers], y[outliers], color='r')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random dataset with some outliers
np.random.seed(1)
data = np.concatenate([np.random.normal(0, 1, 100), np.random.normal(
    10, 1, 20), np.random.normal(20, 1, 5)])

# Create a box plot of the dataset
fig, ax = plt.subplots()
ax.boxplot(data)

# Add a title and axis labels
ax.set_title('Dataset with Outliers')
ax.set_xlabel('Data')
ax.set_ylabel('Values')

# Calculate the quartiles and interquartile range
q1, q2, q3 = np.percentile(data, [25, 50, 75])
iqr = q3 - q1

# Add the quartiles and IQR to the plot
ax.axhline(q2, color='k', linestyle='-', label='Median')
ax.axhline(q1, color='k', linestyle='-.', label='Q1')
ax.axhline(q3, color='k', linestyle='--', label='Q3')
ax.text(0.95, q2, f" IQR = {iqr:.2f}",
        transform=ax.get_yaxis_transform(), ha='right')

# Calculate the upper and lower bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Mark any outliers outside of the bounds
outliers = [x for x in data if x < lower_bound or x > upper_bound]
ax.scatter(np.repeat(1, len(outliers)), outliers,
           color='r', marker='x', zorder=10)

# Add labels for the quartiles, median, and outliers
ax.text(1.05, q2, f"Median: {q2:.2f}", transform=ax.get_yaxis_transform(), va='center')
ax.text(1.05, q1, f"Q1: {q1:.2f}", transform=ax.get_yaxis_transform(), va='center')
ax.text(1.05, q3, f"Q3: {q3:.2f}", transform=ax.get_yaxis_transform(), va='center')
ax.text(1.1, upper_bound, f"Upper Bound: {upper_bound:.2f}", color='r', va='center')
ax.text(1.1, lower_bound, f"Lower Bound: {lower_bound:.2f}", color='r', va='center')
ax.text(1.15, np.mean(outliers), f"Outliers: {len(outliers)}", color='r', va='center')

# plot style
plt.style.use("seaborn")

# Show the plot
ax.legend()
plt.show()
```

## **Detecting outlier using Z score**

### Using Z score

$$z = \frac{X - \mu}{\sigma}$$

Formula for Z score = (Observation  Mean)/Standard Deviation


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13,
           12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]


def detect_outliers(data):
    outliers = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)

    return outliers


outlier_pts = detect_outliers(dataset)
print(outlier_pts)
```

```python
# Ploting the z-score points
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13,
           12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]


def detect_outliers(data):
    outliers = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)

    z_scores = []  # create a new list to store z-scores

    for i in data:
        z_score = (i - mean) / std
        z_scores.append(z_score)  # append the z-score to the list
        if np.abs(z_score) > threshold:
            outliers.append(i)
    
    # plot style
    plt.style.use("seaborn")

    # plot the z-scores against the index
    plt.plot(range(len(data)), z_scores, 'bo')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of Dataset')
    plt.show()

    return outliers

outlier_pts = detect_outliers(dataset)
print(outlier_pts)
```

### InterQuantile Range

`75%` - `25%` values in a dataset

**Steps:**
1. Arrange the data in increasing order
2. Calculate first $(q_{1})$ and third quartile $(q_{3})$
3. Find interquartile range (q3-q1)$(q_{3} - q_{1})$
4. Find lower bound $q_{1}\times1.5$
5. Find upper bound $q_{3}\times1$.
   
Anything that lies outside of lower and upper bound is an outlier

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Define the dataset
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13,
           12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]

# Sort the dataset
sorted_dataset = sorted(dataset)

# Calculate the 25th and 75th percentiles (Q1 and Q3)
quantile_1, quantile_3 = np.percentile(sorted_dataset, [25, 75])

print("Q1 (25th percentile):", quantile_1)
print("Q3 (75th percentile):", quantile_3)

# Calculate the interquartile range (IQR)
iqr = quantile_3 - quantile_1
print("IQR:", iqr)

# Calculate the lower bound and upper bound values
lower_bound_val = quantile_1 - (1.5 * iqr)
upper_bound_val = quantile_3 + (1.5 * iqr)

print("Lower Bound Value:", lower_bound_val)
print("Upper Bound Value:", upper_bound_val)
```

```python
import numpy as np

# Create a NumPy array containing the data
data = np.array([3, 7, 8, 11, 15, 19, 21, 24, 25, 27, 29, 30, 32])

# Calculate the 25th percentile (Q1), 50th percentile (Q2 or median), and 75th percentile (Q3) using np.quantile
q1, q2, q3 = np.quantile(data, [0.25, 0.5, 0.75])

# Calculate the interquartile range (IQR) by subtracting Q1 from Q3
iqr = q3 - q1

# Calculate the upper bound by adding 1.5 times the IQR to Q3
upper_bound = q3 + 1.5 * iqr

# Calculate the lower bound by subtracting 1.5 times the IQR from Q1
lower_bound = q1 - 1.5 * iqr

# Print the calculated values
print("IQR:", iqr)
print("Upper Bound:", upper_bound)
print("Lower Bound:", lower_bound)
```
